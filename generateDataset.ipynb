```python
import pandas as pd
import numpy as np

# For reproducibility
np.random.seed(42)

n = 100

# Generate random values
age = np.random.randint(0, 100, n)
income = np.random.randint(0, 100001, n)

# Approval logic
approval = []

for a, inc in zip(age, income):
    if a < 18 or a > 65:
        approval.append("Reject")
    elif 18 <= a <= 65 and 33001 <= inc <= 100000:
        approval.append("Approve")
    else:
        approval.append("Reject")

# Create DataFrame
df = pd.DataFrame({
    "Age": age,
    "Income": income,
    "Loan_Approval": approval
})

# Save to CSV
df.to_csv("dataset_loan.csv", index=False)
```
